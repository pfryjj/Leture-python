```python
class Student:
    def __init__(self, name, dept, mid, final):
        self.name = name
        self.dept = dept
        self.mid = mid
        self.final = final

    def __str__(self):
        return "학과: " + self.dept + ", 이름: " + self.name + ", 중간: " + str(self.mid) + ", 기말: " + str(self.final)


    def grade(self):
        total = self.mid + self.final
        average = total / 2
        
        if average >= 90:
            return 'A'
        elif average >= 80 and average < 90:
            return 'B'
        elif average >= 70 and average < 80:
            return 'C'
        elif average >= 60 and average < 70:
            return 'D'
        else:
            return 'F'

student1 = Student('김경철', '기계학과', 89, 90)

print(student1)
print("학점: " + student1.grade())
        
```

```python
class Car:
    def __init__(self, name, year, color):
        self.name = name
        self.year = year
        self.color = color
    def __str__(self):
        return "자동차 회사: " + self.name + " 년식: " + str(self.year) + " 색상 " + self.color
    def __eq__(self, other):
        if self.name == other.name and self.year == other.year and self.color == other.color:
            return True
        else:
            return False
mycar = Car('현대', '2020', '검정')
yourcar = Car('기아', '2021', '백색')
print(mycar)
print(yourcar)
print(mycar == yourcar)
```
