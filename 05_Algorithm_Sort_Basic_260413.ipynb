```python
#정수 N을 받아서 base에 대한 집법의 수로 변환
def convert_base(N, base):
    #N을 base로 나눈 몫을 몫이 0이 아닌 동안 계속 N으로 나눈다.
    #나눈 나머지를 진법에 해당하는 숫자를 변환하여 리스트에 저장한다.
    #N : 몫을 저장
    #new_list : 나머지를 저장하는 리스트
    new_list = ''
    digits = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    while N != 0:
        new_list = digits[N % base]+ new_list
        N //= base
    return new_list
print(convert_base(26,16))
```

```python
# selection sort
# 정렬이 안된 부분 (start 이후)에서 가장 작은 수를 찾아서 
# start에 있는 수와 swap
# start += 1
def print_step(step, A):
    print(f'{step:2} 단계 : {A}')

def select(A, start):
    min_value = A[start]
    min_index = start
    
    for i in range(start+1, len(A)):
        if A[i] < min_value: 
            min_value = A[i]
            min_index = i
            
    return min_index
            
    return min_index
    
def selection_sort(data_list):
    for start in range(len(data_list)-1):
        min_index = select(data_list, start)
        data_list[start], data_list[min_index] = data_list[min_index], data_list[start]
A = [23, 5, 44, 1, 6]
print('원래 데이터 :', A)
selection_sort(A)
print('정렬 후 데이터 : ',A)
```

```python
def insertion_sort(A):
    for cur_index in range(1, len(A)):
        key = A[cur_index]   # 수정
        
        i = cur_index - 1
        
        while i >= 0 and A[i] > key:   # 수정
            A[i+1] = A[i]
            i -= 1
            
        A[i+1] = key
        
        print_step(cur_index, A)
```

```python
def bubble_sort(A):
    for last in range(len(A)-1, 0, -1):
        is_changed = False
        for i in range(last):
            if A[i] > A[i+1]:
                A[i], A[i+1] = A[i+1], A[i]
                is_changed = True
        print_step(last, A)
        if not is_changed: return
```

```python
A = [1,23, 5, 14, 21, 44, 1, 6]
print('원래 데이터 :', A)
#insertion_sort(A)
bubble_sort(A)
print('정렬 후 데이터 : ',A)
```
