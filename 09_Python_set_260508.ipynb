```python
class circle:
    pi = 3.14
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return self.radius ** 2 * circle.pi
c = circle(2.5)
print(c.area())

```

```python
class Rectangle:

    def __init__(self, width, height):

        self.width = width

        self.height = height



    def area(self):

        return self.width * self.height



rec = Rectangle(2.1, 3.4)



print(rec.area())
```

```python
class Account:
    def __init(self):  # 🚨 문제 1: 언더바가 뒤에 빠졌어요!
        self.balance = 0
    def deposit(self, amount):
        self.balance += amount
        self.balance -= amount  # 🚨 문제 2: 입금했는데 왜 바로 출금을 할까요?

a1 = Account()
a1.deposit(100)
print(a1.balance)
```

```python
class Account:
    # 1. 언더바 양쪽에 잘 붙이기 (__init__)
    def __init__(self):
        self.__balance = 0  # 잔고 숨기기
        
    # 2. 입금할 때도 숨겨진 변수(__balance) 사용하기
    def deposit(self, amount):
        self.__balance += amount
        
    # 3. 출금하기
    def withdraw(self, amount):
        self.__balance -= amount
        
    # 4. 숨겨진 잔고 확인하기
    def get_balance(self):
        return self.__balance

# 5. 들여쓰기 밖으로 빼기 (클래스 밖에서 실행)
a1 = Account()
a1.deposit(100)

a2 = Account()
a2.deposit(50)

print("a1 : ", a1.get_balance())
print("a2 : ", a2.get_balance())
    
```
