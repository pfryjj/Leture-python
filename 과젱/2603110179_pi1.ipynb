```python
import math
import random

def getare(r):
    return r * r * math.pi
radius = random.randint(1, 10)
area = round(getare(radius), 2)

print(f"원의 반지름 {radius}")
print(f"원주율 pi: {math.pi}")
print(f"반지름 {radius}인 원의 넓이: {area}")
```

```python
#3번 문제
import math
import statistics

if __name__ == '__main__':
    
    for i in range(1, 17, 5):
        print(f"{i}! = {math.factorial(i)}")
        
    print()
    st = [80, 99, 77, 65, 92, 74, 82]
    
    print(st)
    print(f"중앙 값: {statistics.median(st):.2f}")
    print(f"평균: {statistics.mean(st):.2f}")
    print(f"분산: {statistics.variance(st):.2f}")
    print(f"표준편차: {statistics.stdev(st):.2f}")
```
